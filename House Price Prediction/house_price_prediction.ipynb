Importing Dependencies

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import  XGBRegressor
```

Importing dataset

```python
house_price_data = pd.read_csv("BostonHousing.csv")
house_price_data.head()
```

Dataset Analysing

```python
house_price_data.describe()
```

```python
house_price_data.shape
```

```python
house_price_data.isnull().sum()
```

Understanding correlation between various features in dataset

```python
correlation =house_price_data.corr()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(correlation,cbar=True,square=True,annot=True,cmap='Blues') 
```

Spliting the data and target

```python
X = house_price_data.drop(columns='price',axis=1)
Y = house_price_data['price']
```

```python
print(X)
```

```python
print(Y)
```

plitting the data into training and testing

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X_train.shape,X_test.shape)
```

Model Training

```python
model =XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

Evaluation

```python
#prediction on training data
X_train_prediction = model.predict(X_train)
print(X_train_prediction)
```

```python
#R squared error 
score_1 = metrics.r2_score(Y_train,X_train_prediction)

#Mean Absoute Error
score_2 =  metrics.mean_absolute_error(Y_train,X_train_prediction)

print("R squared error : ",score_1)
print("Mean Absolute Error :",score_2)
```

Prediction on Test Data

```python
X_test_prediction=model.predict(X_test)
```

```python
#R squared error 
score_1 = metrics.r2_score(Y_test,X_test_prediction)

#Mean Absoute Error
score_2 =  metrics.mean_absolute_error(Y_test,X_test_prediction)

print("R squared error : ",score_1)
print("Mean Absolute Error :",score_2)
```

Visualize the actual Prices and predicted

```python
plt.scatter(Y_train,X_train_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual price vs Predicted price")
plt.show()
```
